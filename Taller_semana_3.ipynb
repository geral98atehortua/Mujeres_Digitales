## Ejercicio 3

Se le entregan los datos de movimientos de acciones en el archivo ```DUK.csv```, y se le pide que entregue el promedio de Open del 2015, es decir entre las fechas 2015-01-01 y 2015-12-31

````python
# Escriba su código aquí
import pandas as pd
from google.colab import files
uploaded = files.upload()
````

````python
import io
df =pd.read_csv(io.BytesIO(uploaded['DUK.csv']), sep=",")
df.head(1259)
````

````python
df = df[(df['Date']>='2015-01-01')& (df['Date']<='2015-12-31')]
df.iloc[0:1259]  
df.iloc[:, 0:2] 
````

````python
p=df.Open.mean()
print('el promedio de open entre las fechas 2015-01-01 y 2015-12-31 es: ',round(p,5))
````
